```python
import pandas as pd
import numpy as np
```

```python
# Load dataset
df = pd.read_excel("studentperformance.xlsx")
```

```python
# Display first 5 rows
print("Dataset Preview:")
print(df.head())
```

```python
# Check missing values
print("\nMissing Values:")
print(df.isnull().sum())
```

```python
# Select travel time column
# (Change column name if needed)
travel = df['traveltime']
```

```python
travel
```

```python
# Mean of travel time
mean_value = travel.mean()
print("\nMean of Travel Time:", mean_value)
```

```python
# Fill missing values with mean
df_mean = df.copy()
df_mean['traveltime'] = df_mean['traveltime'].fillna(mean_value)
```

```python
df_mean['traveltime']
```

```python
# Median of travel time
median_value = travel.median()
print("Median of Travel Time:", median_value)
```

```python
# Fill missing values with median
df_median = df.copy()
df_median['traveltime'] = df_median['traveltime'].fillna(median_value)
```

```python
# Compare means
print("\nBefore Replacement Mean:", travel.mean())
print("After Mean Replacement:", df_mean['traveltime'].mean())
print("After Median Replacement:", df_median['traveltime'].mean())
```

```python
# Compare medians
print("\nBefore Replacement Median:", travel.median())
print("After Mean Replacement Median:", df_mean['traveltime'].median())
print("After Median Replacement Median:", df_median['traveltime'].median())
```
