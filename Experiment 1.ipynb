## STEP 1 : Import & Load Data

```python
import pandas as pd

df = pd.read_csv("retail_data.csv")
```

## STEP 2 : Checking Data Shape and Columns and data types

```python
print(df.shape)      # rows and columns
print(df.columns)    # column names
print(df.info())     # data types + missing values
```

```python
df.head()
```

```python
df['category_id'] = df['category_id'].astype('Int64')
df['product_id'] = df['product_id'].astype('Int64')
df['user_id'] = df['user_id'].astype('Int64')
```

```python
df.head()
```

```python
df.columns
```

```python
print("Missing Values BEFORE cleaning:\n")
print(df.isnull().sum())
```

```python
df = df.dropna(subset=['user_id'])
```

```python
df.shape
```

```python
print("Missing Values AFTER cleaning:\n")
print(df.isnull().sum())
```

##  Step 5 : Check Duplicates Rows

```python
print("Missing Values AFTER cleaning:\n")
print(df.isnull().sum())
```

```python
print("Duplicate Rows BEFORE:", df.duplicated().sum())
```

```python
df = df.drop_duplicates()
```

```python
print("Duplicate Rows AFTER:", df.duplicated().sum())
```

## STEP 3 : Handiling Missing Values

```python
cat_cols = df.select_dtypes(include=['object']).columns
```

```python
cat_cols
```

```python
# Categorical → Mode

for col in cat_cols:
    df[col] = df[col].fillna(df[col].mode()[0])
```

```python
df['price'] = df['price'].fillna(df['price'].mean())
```

```python
df['category_id'] = df['category_id'].fillna(0)
df['product_id'] = df['product_id'].fillna(0)
```

## Step 6 : Check Inconsistent Text

```python
print("Unique Country Values BEFORE:\n")
print(df['brand'].unique())
print(df['brand'].nunique())
```

```python
df = df.copy()
# Remove spaces
df['brand'] = df['brand'].str.strip()

# Convert to lowercase
df['brand'] = df['brand'].str.lower()
```

```python
print("Unique Country Values AFTER:\n")
print(df['brand'].unique())
print(df['brand'].nunique())
```

```python
df.head(5)
```

 ## Step 7 : Check Invalid Values

```python
df[df['price']<0]
```

```python
print("Negative Quantity Count:", (df['price'] < 0).sum())
```

```python
df['price'] = df['price'].abs()
```

```python
print("Negative Price Count:", (df['price'] < 0).sum())
```

 ## Step 9: Standardize Column Names

```python
print(df.columns)
```

```python
df.columns = df.columns.str.lower().str.replace(" ", "_")
print(df.columns)
```

## Step 10 : Final Check

```python
print("Final Shape:", df.shape)

print("\nFinal Info:\n")
print(df.info())

print("\nPreview:\n")
print(df.head())
```

## Step 11 : Save Cleaned Dataset

```python
df.isnull().sum()
```

```python
df.to_csv("cleaned_retail_dataset.csv", index=False)

print("✅ Cleaned dataset saved")
```
