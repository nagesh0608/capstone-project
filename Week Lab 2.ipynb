```python
import pandas as pd

df = pd.read_csv("https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv")
df.head()
```

```python
# Dataset info
df.info()

# Check missing values
df.isnull().sum()
```

```python
# Fill Age with median
df['Age'] = df['Age'].fillna(df['Age'].median())

# Fill Embarked with mode
df['Embarked'] = df['Embarked'].fillna(df['Embarked'].mode()[0])

# Drop Cabin (too many missing values)
df = df.drop(columns=['Cabin'])

# Check again
df.isnull().sum()
```

```python
df['Age_Group'] = pd.cut(df['Age'],
                        bins=[0, 18, 35, 60, 100],
                        labels=['Child', 'Young', 'Adult', 'Senior'])
```

```python
df['Family_Size'] = df['SibSp'] + df['Parch'] + 1
```

```python
df['Title'] = df['Name'].str.extract(' ([A-Za-z]+)\.', expand=False)
```

```python
import numpy as np
```

```python
df['Fare_Log'] = np.log1p(df['Fare'])
```

```python
# Min-Max Normalization
df['Age_Norm'] = (df['Age'] - df['Age'].min()) / (df['Age'].max() - df['Age'].min())

df['Fare_Norm'] = (df['Fare'] - df['Fare'].min()) / (df['Fare'].max() - df['Fare'].min())
```

```python
df['Sex'] = df['Sex'].map({'male': 0, 'female': 1})
```

```python
pd.set_option('display.max_columns', None)
df.head()
```

```python
#df = pd.get_dummies(df, columns=['Embarked', 'Age_Group', 'Title'], drop_first=True)
df = pd.get_dummies(df, columns=['Embarked', 'Age_Group', 'Title'], drop_first=True, dtype=int)
```

```python
df.columns
```

```python
df = df.drop(columns=['Name', 'Ticket'])
```
