```python
# calculating quartiles using percentiles
```

```python
q1 = np.percentile(likes,25)
```

```python
print(q1)
```

```python
q2= np.percentile(likes,50)
print(q2)
```

```python
q3 = np.percentile(likes,75)
print(q3)
```

```python
iqr = q3 - q1
print(iqr)
```

```python
# detecting outlies in likes arrary
```

```python
lower_bound = q1 - 1.5*iqr
upper_bound = q3 + 1.5*iqr
print(lower_bound,upper_bound)
```

```python
#  likes = [ 100,120,140,160,160,160,180,200,200,200,220,250,300,5000]
```

```python
outliers = []

for i in likes:
    if i < lower_bound or i > upper_bound :
        outliers.append(i)

print(outliers)
        
```
