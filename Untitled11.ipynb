## Data Quality and Cleaning

```python
# (a) Are there any missing values?
df.isnull().sum()
```

```python
# Ans: No
```

```python
# (b) Any invalid age values (outside 15–22)?
df[(df['age'] < 15) | (df['age'] > 22)]
```

```python
# Ans No
```

```python
# (c) Are there duplicate records?
print(df.duplicated().sum())
```

```python
# No
```

```python
# (d) Are G1, G2, G3 scores valid (0–20)?
df[(df['G1'] < 0) | (df['G1'] > 20) |
   (df['G2'] < 0) | (df['G2'] > 20) |
   (df['G3'] < 0) | (df['G3'] > 20)]
```

```python
# Ans:  All values are valid
```

```python
# (e) Completeness percentage of dataset
print((df.notnull().sum().sum() / df.size) * 100)
```
