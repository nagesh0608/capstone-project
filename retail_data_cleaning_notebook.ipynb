
# Data Quality & Cleaning

 ## Missing Values


$$
\text{Missing Values} = \sum_{i=1}^{n} I(x_i = \text{NULL}),
\quad
I(x_i) =
\begin{cases}
1, & \text{if } x_i \text{ is NULL} \\
0, & \text{if } x_i \text{ is not NULL}
\end{cases}
$$

```python
#a
df.isnull().sum()
```

###  Completeness

$$
\text{Completeness} = \frac{\text{Non-missing}}{\text{Total}} \times 100
$$

```python
#e
total_cells = df.size
missing_cells = df.isnull().sum().sum()
((total_cells - missing_cells) / total_cells) * 100
```

### Accuracy

$$
\text{Accuracy} = \frac{\text{correct records}}{\text{Total records}} \times 100
$$

```python
#f
valid = df[df['price'] > 0]
(len(valid) / len(df)) * 100
```

###  Consistency

$$
\text{Consistency} = \frac{\text{Consistent records}}{\text{Total records}} \times 100
$$

```python
#g
valid_events = ['view','cart','purchase']
consistent = df[df['event type'].isin(valid_events)]
(len(consistent) / len(df)) * 100
```
