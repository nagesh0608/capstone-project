# 📘 Data Sources in Data Science

## 🟦 What are Data Sources?
Data sources are places from which data is collected for analysis.

## 📂 Common Data Sources
- Databases (SQL, NoSQL)
- Web data (APIs, scraping)
- Sensors & IoT devices
- Social media platforms
- Transactional systems (sales, banking)


## 🟩 Databases (SQL & NoSQL)

### 📌 Explanation
Databases are the most common and reliable source of structured and semi-structured data.

- SQL → Tables (rows & columns)
- NoSQL → JSON, key-value, documents

### 📌 Examples
- College student database
- E-commerce product & orders
- Banking systems


```python
# Example: SQL-like structured data using pandas
import pandas as pd

data = {
    "ID": [1, 2, 3],
    "Name": ["Ravi", "Sita", "Kumar"],
    "Marks": [85, 78, 92]
}

df = pd.DataFrame(data)
print(df)
```

## 🌐 Web Data (APIs & Web Scraping)

### 📌 Explanation
- APIs → Structured data access
- Web scraping → Extract from websites

### 📌 Examples
- Weather data
- Stock prices
- Product prices


```python
# Example: simple API request
import requests

url = "https://api.github.com"
response = requests.get(url)

print("Status Code:", response.status_code)
print("Data:", response.json())
```

## 📡 Sensors & IoT Devices

### 📌 Explanation
Devices collect real-time streaming data.

### 📌 Examples
- Temperature sensors
- Fitness trackers
- Smart agriculture


```python
# Example: simulated sensor data
import random

temperature = random.uniform(20, 35)
print("Temperature:", round(temperature, 2))
```

## 📱 Social Media Platforms

### 📌 Explanation
Generate large unstructured and semi-structured data.

### 📌 Examples
- Posts, comments
- Reviews
- Hashtags


```python
# Example: simple text analysis
text = "This product is very good and useful"

words = text.split()
print("Words:", words)
print("Total words:", len(words))
```

## 💳 Transactional Systems

### 📌 Explanation
Record business operations.

### 📌 Examples
- Sales transactions
- Banking transactions
- Ticket booking


```python
# Example: transaction data
transactions = [
    {"item": "Milk", "price": 50},
    {"item": "Bread", "price": 30}
]

total = sum(t["price"] for t in transactions)
print("Total Sales:", total)
```

## 🟦 Summary

- Databases → Structured data  
- Web data → Real-time data  
- Sensors/IoT → Streaming data  
- Social media → Unstructured data  
- Transactional → Business data
