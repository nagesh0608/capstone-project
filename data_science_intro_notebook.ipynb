# 📊 Data Science
### An Introduction

## 📌 What is Data Science?

Data Science is an interdisciplinary field that uses:
- Scientific methods
- Algorithms
- Systems

to extract **knowledge and insights from data**.

---
### It combines:
- 📈 Statistics
- 💻 Computer Science
- 🧠 Domain Knowledge

## 🎯 Key Goals of Data Science

- 🔍 Extract meaningful insights
- 🧾 Support decision-making
- 🔮 Predict future trends

## 📂 Types of Data

### a) Structured Data
- Organized in **tabular format (rows & columns)**
- Easy to store and analyze

### Example:
- Databases
- Spreadsheets

```python
# Structured data example using pandas
import pandas as pd

data = {
    'Name': ['A', 'B', 'C'],
    'Marks': [90, 80, 70]
}

df = pd.DataFrame(data)
print(df)
```

## 📂 Types of Data (Continued)

### b) Unstructured Data
- No predefined format
- Difficult to process directly

### Example:
- Text
- Images
- Videos
- Social media posts

```python
# Unstructured data example
text = "Data Science is powerful!"
print(text)
```

## 🔍 Simple Data Analysis Example

```python
# Basic analysis
print(df.head())
print(df.describe())
```

## 📈 Simple Visualization Example

```python
import matplotlib.pyplot as plt

plt.bar(df['Name'], df['Marks'])

plt.title("Marks of Students")
plt.xlabel("Name")
plt.ylabel("Marks")

plt.show()
```
