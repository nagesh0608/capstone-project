```python
import numpy as np
import pandas as pd
```

```python
likes = [100, 120, 140, 160, 180, 200, 220, 250, 300, 5000]
```

```python
# Question 2: Find the mean of the dataset
print(np.mean(likes))
```

```python
# Question 3: Find the median of the 
print(np.median(likes))
```

```python
df = pd.DataFrame(likes,columns=['likes'])
```

```python
df
```

```python
# Question 4: Find the mode of the datd
df['likes'].mode()
```

```python
# Question 5: Find the range of the dataset
range_value = max(likes)-min(likes)
```

```python
range_value
```

```python
# Question 6: Find the variance of the dataset
print(np.var(likes))
```

```python
# Question 7: Find the standard deviation of the dataset
print(np.std(likes))
```

```python
# Question 8: Find the quartiles (Q1, Q2, Q3)
q1 = np.percentile(likes,25)
q2 = np.percentile(likes,50)
q3 = np.percentile(likes,75)
```

```python
print(q1,q2,q3)
```

```python
# Question 10: Detect the outliers using IQR method
IQR = q3-q1
```

```python
IQR
```

```python
lower_bound = q1 - 1.5 * IQR
upper_bound = q3 + 1.5 * IQR
```

```python
outliers =[]

for x in likes:
    if x < lower_bound or x >upper_bound:
        outliers.append(x)

print(outliers)
```

# how do affect outliers to stability
Ans: Outliers affect stability because they pull results away from the normal pattern.
