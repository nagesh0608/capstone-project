```python
import pandas as pd
```

```python
# Load dataset
df = pd.read_excel("studentperformance.xlsx")
```

```python
# Check missing values
print(df.isnull().sum())
```

```python
# Mean of traveltime
mean_value = df['traveltime'].mean()
print("Mean:", mean_value)
```

```python
# Replace missing values with mean
df['traveltime'] = df['traveltime'].fillna(mean_value)
```

```python
# Median of traveltime
median_value = df['traveltime'].median()
print("Median:", median_value)
```

```python
# Replace missing values with median
df['traveltime'] = df['traveltime'].fillna(median_value)
```

Comment:

Mean replacement keeps the average nearly same.
Median replacement is better when outliers are present.
Data becomes complete after filling missing values.
