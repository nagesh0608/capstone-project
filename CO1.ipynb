# 📘 Data Science – Introduction

## 🟦 What is Data Science?
Data Science is an interdisciplinary field that uses scientific methods, algorithms, and systems to extract knowledge and insights from data.

Combines:
- Statistics
- Computer Science
- Domain Knowledge

## 🎯 Key Goals
- Extract meaningful insights
- Support decision-making
- Predict future trends

```python
# Simple example: average marks
marks = [85, 78, 92, 88, 76]
average = sum(marks) / len(marks)
print("Average Marks:", average)
```

## 🟩 Structured Data
Organized in rows & columns.
Easy to analyze.

| Student_ID | Name  | Marks | Grade |
|------------|-------|-------|-------|
| 101 | Ravi | 85 | A |
| 102 | Sita | 78 | B |
| 103 | Kumar | 92 | A |

```python
import pandas as pd

data = {
    "Student_ID": [101, 102, 103],
    "Name": ["Ravi", "Sita", "Kumar"],
    "Marks": [85, 78, 92],
    "Grade": ["A", "B", "A"]
}

df = pd.DataFrame(data)
print(df)
```

## 🟥 Unstructured Data
No fixed format.
Examples: text, images, videos

```python
text = "Data Science is an evolving field"
words = text.split()
print(words)
print("Word Count:", len(words))
```

## 🟨 Semi-Structured Data
Organized using keys/tags (JSON, XML)

```python
import json

json_data = '{"Student_ID":101, "Name":"Ravi", "Marks":85, "Grade":"A"}'
data = json.loads(json_data)

print(data)
print("Student Name:", data["Name"])
```

## 🟦 Summary
- Structured → Tables
- Semi-structured → JSON/XML
- Unstructured → Text, images

```python
# Identify data types
import pandas as pd

data1 = {"Name": "Ravi", "Marks": 85}
data2 = "This is a text document"
data3 = pd.DataFrame({"Marks": [85, 90]})

print(data1)
print(data2)
print(data3)
```
