# 📘 Data Science – Introduction

## 🟦 What is Data Science?
Data Science is an interdisciplinary field that uses scientific methods, algorithms, and systems to extract knowledge and insights from data.

Combines:
- Statistics
- Computer Science
- Domain Knowledge

The formula for RMSE is $RMSE = \sqrt{\frac{1}{n} \sum (y_i - \hat{y}_i)^2}$

## 🎯 Key Goals
- Extract meaningful insights
- Support decision-making
- Predict future trends

````python
# Simple example: average marks
marks = [85, 78, 92, 88, 76]
average = sum(marks) / len(marks)
print("Average Marks:", average)
````

## 🟩 Structured Data
Organized in rows & columns.
Easy to analyze.

| Student_ID | Name  | Marks | Grade |
|------------|-------|-------|-------|
| 101 | Ravi | 85 | A |
| 102 | Sita | 78 | B |
| 103 | Kumar | 92 | A |

````python
import pandas as pd

data = {
    "Student_ID": [101, 102, 103],
    "Name": ["Ravi", "Sita", "Kumar"],
    "Marks": [85, 78, 92],
    "Grade": ["A", "B", "A"]
}

df = pd.DataFrame(data)
print(df)
````

## 🟨 Semi-Structured Data

### 📌 Definition
Semi-structured data is data that does not follow a strict tabular format but contains some level of organization using:

- Keys
- Tags
- Markers

### 📂 Examples
- JSON
- XML
- HTML

### 📌 Example (JSON)

```json
{
  "Student_ID": 101,
  "Name": "Ravi",
  "Marks": 85,
  "Grade": "A"
}
```

👉 Not in table form  
👉 Organized using key–value pairs


````python
# Semi-structured data example using JSON

import json

json_data = '''
{
  "Student_ID": 101,
  "Name": "Ravi",
  "Marks": 85,
  "Grade": "A"
}
'''

data = json.loads(json_data)

print("Complete Data:", data)
print("Student Name:", data["Name"])
print("Marks:", data["Marks"])
````

## 🟦 Summary
- Structured → Tables
- Semi-structured → JSON/XML
- Unstructured → Text, images
