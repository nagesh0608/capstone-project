```python
import matplotlib.pyplot as plt

# Data
x = [1, 2, 3, 4, 5]

y1 = [2, 4, 6, 8, 10]
y2 = [1, 3, 5, 7, 9]
y3 = [2, 5, 3, 7, 4]

# Plot Line 1 (fully styled)
plt.plot(x, y1,
         color='blue',            # line color
         linestyle='-',           # solid line
         linewidth=2,             # thickness
         marker='o',              # marker type
         markersize=8,            # marker size
         markerfacecolor='yellow',# inside color
         markeredgecolor='black', # border color
         alpha=0.9,               # transparency
         label='Linear Growth')   # legend label

# Plot Line 2
plt.plot(x, y2,
         color='red',
         linestyle='--',
         linewidth=2,
         marker='s',
         markersize=8,
         markerfacecolor='white',
         markeredgecolor='black',
         alpha=0.8,
         label='Odd Pattern')

# Plot Line 3
plt.plot(x, y3,
         color='green',
         linestyle=':',
         linewidth=2,
         marker='^',
         markersize=8,
         markerfacecolor='pink',
         markeredgecolor='black',
         alpha=0.7,
         label='Non-Linear')

# Title and labels
plt.title("Complete Styled Multiple Line Graph", fontsize=14)
plt.xlabel("X Axis", fontsize=12)
plt.ylabel("Y Axis", fontsize=12)

# Axis limits
plt.xlim(0, 6)
plt.ylim(0, 12)

# Grid
plt.grid(True, linestyle='--', alpha=0.5)

# Legend
plt.legend(loc='upper left', fontsize=10)

# Show plot
plt.show()
```

```python
import numpy as np

# Python list
list1 = [20, 30, 40, 50]

# NumPy array
arr = np.array([20, 30, 40, 50])

# 🔸 Update single value
list1[1] = 100
arr[1] = 100

print("List:", list1)
print("Array:", arr)


# 🔸 Update multiple values
list1[1:3] = [200, 300]
arr[1:3] = [200, 300]

print("List:", list1)
print("Array:", arr)


# 🔸 Conditional update
# List (needs loop)
for i in range(len(list1)):
    if list1[i] == 40:
        list1[i] = 400

# NumPy (direct)
arr[arr == 40] = 400

print("List:", list1)
print("Array:", arr)
```

```python
# ==============================
# 1. Python automatic datatype
# ==============================

x = 10          # Python assigns int automatically
y = 10.5        # Python assigns float automatically
z = "Hello"     # Python assigns string

print(type(x))  # int
print(type(y))  # float
print(type(z))  # str


# ==============================
# 2. NumPy automatic datatype
# ==============================

import numpy as np

a = np.array([1, 2, 3])   # NumPy assigns datatype automatically
print(a.dtype)            # usually int64


# ==============================
# 3. NumPy manual datatype
# ==============================

b = np.array([1, 2, 3], dtype=np.int32)  # force datatype
print(b.dtype)                           # int32


# ==============================
# 4. Large number → automatic int64
# ==============================

c = np.array([10000000000])  # large number
print(c.dtype)               # int64 (needs more memory)


# ==============================
# 5. Pandas automatic datatype
# ==============================

import pandas as pd

df = pd.DataFrame({
    'A': [1, 2, 3],          # numeric → int64
    'B': [10.5, 20.2, 30.1], # float → float64
    'C': ['x', 'y', 'z']     # text → object
})

print(df.dtypes)


# ==============================
# 6. Pandas manual datatype change
# ==============================

df['A'] = df['A'].astype('int32')  # convert int64 → int32
print(df.dtypes)


# ==============================
# 7. Memory difference
# ==============================

arr1 = np.array([1,2,3], dtype=np.int32)
arr2 = np.array([1,2,3], dtype=np.int64)

print(arr1.nbytes)  # less memory
print(arr2.nbytes)  # more memory
```
